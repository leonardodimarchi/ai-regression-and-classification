# Dataset

O dataset escolhido foi o [House Sales in King County, USA](https://www.kaggle.com/datasets/harlfoxem/housesalesprediction/data). Ele possui preços de casas a venda entre os anos de 2014 e 2015 em King County nos estados unidos.

Iremos observar as colunas para entender o que cada uma significa, com a ajuda do fórum de discussões do Kaggle para as colunas menos intuitivas.

### Atributos

- **ID**: Identificador da casa
- **Date**: Data da venda
- **Price**: Preço da casa em dólares
- **Bedrooms**: Quantidade de quartos
- **Bathrooms**: Quantidade de banheiros (valores 0.5 significam lavabos)
- **Sqft_living**: Metros quadrados da casa
- **Sqft_lot**: Metros quadrados do lote
- **Floors**: Quantidade de andares da casa
- **Waterfront**: Diz se a casa está a beira-mar (0 e 1)
- **View**: Indica a qualidade da vista da propriedade, de 0 a 4.
- **Condition**: Condição atual da casa, de 1 a 5. Quanto maior, melhor conservada
- **Grade**:  Nota da casa após construção, de 1 a 13. Quanto maior, melhor
- **Sqft_above**: Metros quadrados considerando andares que estão acima do térreo
- **Sqft_basement**: Metros quadrados do porão
- **Yr_built**: Ano de construção
- **Yr_renovated**: Ano da última renovação
- **Zipcode**: CEP
- **Lat**: Latitude
- **Long**: Longitude
- **Sqft_living15**: Média de metros quadrados da casa dos 15 vizinhos mais próximos
- **Sqft_lot15**: Média de metros quadrados do lote dos 15 vizinhos mais próximos

```python
import pandas as pd

# Lendo o dataset
dataset_original = pd.read_csv('kc_house_data.csv')

# Exibindo as 5 primeiras linhas do dataset
display(dataset_original.head())

# Exibindo algumas estatísticas dos atributos
dataset_original.describe()
```

# Análise Exploratória de Dados (EDA)

### Verificar se existem valores ausentes 

Não existem valores nulos em nenhum dos atributos deste dataset.

### Transformação dos dados

Transformaremos os seguintes atributos já na etapa de EDA para facilitar a visualização de correção:

- **Date**: vamos separar a data em três novos atributos, ano, mes e dia. 

### Correlação

Plotamos o heatmap da correlação dos atributos para identificar quais são mais relevantes para determinar o preço da casa. 
Removeremos alguns desses atributos na próxima etapa.




```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

dataset = dataset_original.copy()

print("\n\n=========Verificando se existem valores ausentes=========")
soma_de_nulos_por_atributo = dataset.isnull().sum();
print(soma_de_nulos_por_atributo)

print("\n\n=========Transformação dos dados=========")

print("\n- Separando data em ano/mes/dia")
dataset['date'] = pd.to_datetime(dataset['date'])
dataset['year'] = dataset['date'].dt.year
dataset['month'] = dataset['date'].dt.month
dataset['day'] = dataset['date'].dt.day
dataset = dataset.drop('date', axis=1)

print("\n\n=========Correlação=========")

mapa_de_correlacao = dataset.corr()
plt.figure(figsize=(20,15))
sns.heatmap(mapa_de_correlacao, annot=True)
plt.show()
sns.pairplot(dataset, x_vars=['sqft_living','sqft_above','grade'], y_vars='price', height=7, aspect=0.7, kind='reg')
```

# Modelos de regressão

### Regressão linear

```python
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing

# Separando o target e os outros atributos
target = dataset_removed_columns['price']
X = dataset_removed_columns.drop(['price'], axis = 1)

# Separando os dados entre treino e teste
X_train, X_test, y_train, y_test =  train_test_split(X, target, random_state=42, train_size=0.7, shuffle=True)

# Criando normalizador
normalizer = preprocessing.Normalization()
normalizer.adapt(X_train)

# Gerando a visualizando um resumo do modelo
model = keras.Sequential([
    normalizer,
    layers.Dense(1)
])

model.summary()

# Configurando o modelo para utilizar o Adam durante gradiente descendente, utilizando erro quadrático médio para visualizar o loss
model.compile(optimizer=tf.optimizers.Adam(learning_rate=100), loss='mean_squared_error')

# Treino
history = model.fit(
    X_train, y_train,
    epochs=15,
    validation_split = 0.2)

# Plotando o gráfico de loss para treino e teste com base na época
def plot_loss(history):
    plt.plot(history.history['loss'], label='loss')
    plt.plot(history.history['val_loss'], label='val_loss')
    plt.xlabel('Épocas')
    plt.ylabel('Erro')
    plt.legend()
    plt.grid(True)

plot_loss(history)
```

### Árvore de decisão

```python
import sklearn.tree as SKTree
from sklearn.model_selection import train_test_split
import re as re
from sklearn.metrics import r2_score

dataset_tree = dataset_removed_columns.copy()

dataset_tree = dataset_tree.drop(['lat', 'long'], axis=1)

# Separando o target e os outros atributos
target = dataset_tree['price']
X = dataset_tree.drop(['price'], axis = 1)

display(dataset_tree.head())

X_train, X_test, y_train, y_test = train_test_split(X, target, test_size=0.3, random_state=42)

# Historico de treino e teste
history_acc_train = []
history_acc_test = []

for max_depth in range(1,10):
    tree = SKTree.DecisionTreeRegressor(criterion='squared_error', max_depth=max_depth)
    tree = tree.fit(X_train, y_train)

    pred_train = tree.predict(X_train)
    pred_test = tree.predict(X_test)

    r2_train = r2_score(y_train, pred_train)
    r2_test = r2_score(y_test, pred_test)

    history_acc_train.append(r2_train)
    history_acc_test.append(r2_test)

# Avaliando diferentes profundidades para a árvore
plt.plot(history_acc_train)
plt.plot(history_acc_test)
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('max_depth')
plt.legend(['train', 'val'], loc='upper left')
plt.show()

# Testando e plotando a árvore para X de profundidade
tree = SKTree.DecisionTreeRegressor(criterion='squared_error', max_depth=3)
tree = tree.fit(X_train, y_train)

pred_train = tree.predict(X_train)
pred_test = tree.predict(X_test)

r2_train = r2_score(y_train, pred_train)
r2_test = r2_score(y_test, pred_test)

history_acc_train.append(r2_train)
history_acc_test.append(r2_test)

SKTree.plot_tree(tree, fontsize=8)
```
