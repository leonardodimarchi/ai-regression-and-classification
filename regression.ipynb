# Dataset

```python
import pandas as pd

# Lendo o dataset
dataset_original = pd.read_csv('kc_house_data.csv')
```

# Pré-processamento

### Correlação 

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Realizando a cópia do dataset para manter a cópia original 
dataset = dataset_original.copy()

# Convertendo o atributo de data do formato 20141013T000000 para datetime,
# assim representando as datas de maneira numérica para que seja possível calcular a correlação
dataset['date'] = pd.to_datetime(dataset['date'], format='%Y%m%dT%H%M%S')

# Calculando o mapa de correlação
mapa_de_correlacao = dataset.corr()

# Plotando o mapa de correlação
plt.subplots(figsize=(17,14))
sns.heatmap(mapa_de_correlacao, annot=True, linewidths=0.5, linecolor="Black", fmt="1.1f")
plt.title("Correlação dos atributos",fontsize=30)
plt.show()
```

Plotamos o heatmap da correlação dos atributos para identificar quais são mais relevantes para determinar o preço da casa.

### Separação dos dados de treino/teste

```python
from sklearn.model_selection import train_test_split

# Separando o target e os outros atributos
target = dataset_pos_remocao_colunas['price']
X = dataset_pos_remocao_colunas.drop(['price'], axis = 1)

X_train, X_test, y_train, y_test = train_test_split(X, target, test_size = 0.3, random_state=1) 
```

# Modelos

### Regressão linear

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
from sklearn.metrics import r2_score

# Gerando o normalizador
normalizer = preprocessing.Normalization()
normalizer.adapt(X_train)

# Criando o modelo com o normalizador e uma camada com um único perceptron
model = keras.Sequential([
    normalizer,
    layers.Dense(1)
])

# Exibindo um resumo do modelo
model.summary()

# Configurando o modelo para utilizar o Adam durante gradiente descendente, utilizando erro quadrático médio para calculo de loss
model.compile(optimizer=tf.optimizers.Adam(learning_rate=100), loss='mean_squared_error')

# Treino com 100 épocas 
history = model.fit(
    X_train, y_train,
    epochs=100,
    validation_split=0.2,
    verbose=1
)

# Plotando o gráfico de loss para treino e teste com base na época
def plot_loss(history):
    plt.plot(history.history['loss'], label='loss')
    plt.plot(history.history['val_loss'], label='val_loss')
    plt.xlabel('Épocas')
    plt.ylabel('Erro')
    plt.legend()
    plt.grid(True)

plot_loss(history)

# Teste
y_pred = model.predict(X_test)

# Visualizando a métrica R^2, quanto mais próximo de 1, melhor o ajuste do modelo aos dados
r2 = r2_score(y_test, y_pred)
print(f'R-squared (R²): {r2}')

# Plotando gráfico dos valores de teste e dos valores previstos
plt.figure(figsize=(10,6))
plt.scatter(
    y_test,
    y_pred,
    color="blue",
    marker="o"
)
plt.plot(
    y_pred,
    y_pred,
    marker='o',
    color='green',
    markerfacecolor='red',
    markersize=7,
    linestyle='dashed'
)

plt.xlabel("Valores de teste")
plt.ylabel("Valores previstos")
plt.title("Valores previstos / teste")

plt.show()
```

### Árvore de decisão

```python
import sklearn.tree as SKTree
import re as re
from sklearn.metrics import r2_score

# Historico de métricas para treino e teste
history_acc_train = []
history_acc_test = []
profundidade_de_maior_r2 = 1

# Treino e preenchimento do histórico para profundidades de 1 até 15
# Utilizando uma árvore regressora com critério de erro quadrático
for max_depth in range(1,15):
    tree = SKTree.DecisionTreeRegressor(criterion='squared_error', max_depth=max_depth)
    tree = tree.fit(X_train, y_train)

    pred_train = tree.predict(X_train)
    pred_test = tree.predict(X_test)

    r2_train = r2_score(y_train, pred_train)
    r2_test = r2_score(y_test, pred_test)

    history_acc_train.append(r2_train)
    history_acc_test.append(r2_test)

# Exibindo o histórico de métrica R^2 para cada profundidade tanto em treino quanto em teste
plt.plot(history_acc_train)
plt.plot(history_acc_test)
plt.title('Métrica R^2 por profundidade')
plt.ylabel('R^2')
plt.xlabel('Profundidade')
plt.legend(['train', 'val'], loc='upper left')
plt.show()

# Plotando a árvore para uma profundidade escolhida
profundidade = 4

# Exibindo a métrica R^2 para a profundidade escolhida
print(history_acc_train[profundidade])
print(history_acc_test[profundidade])

tree = SKTree.DecisionTreeRegressor(criterion='squared_error', max_depth=profundidade)
tree = tree.fit(X_train, y_train)

pred_train = tree.predict(X_train)
pred_test = tree.predict(X_test)

r2_train = r2_score(y_train, pred_train)
r2_test = r2_score(y_test, pred_test)

history_acc_train.append(r2_train)
history_acc_test.append(r2_test)

SKTree.plot_tree(tree, fontsize=8)
```

### Redes neurais

```python
from sklearn.metrics import accuracy_score
import numpy as np

# Gerando o normalizador
normalizer = preprocessing.Normalization()
normalizer.adapt(X_train.values)

# Criando o modelo com o normalizador e algumas camadas
model = keras.Sequential([
    normalizer,
    layers.Dense(16),
    layers.Dense(16, activation='relu'),
    layers.Dense(16, activation='relu'),
    layers.Dense(16, activation='relu'),
    layers.Dense(8, activation='relu'),
    layers.Dense(4, activation='relu'),
    layers.Dense(1)
])

# Exibindo um resumo do modelo
model.summary()

# Configurando o modelo para utilizar o Adam durante gradiente descendente, utilizando erro quadrático médio para calculo de loss
model.compile(optimizer='adam', loss='mse')

# Treino com 100 épocas
history = model.fit(
    X_train, y_train,
    epochs=100,
    validation_split=0.2,
    verbose=1
)

# Plotando o gráfico de loss para treino e teste com base na época
plot_loss(history)

# Teste
y_pred = model.predict(X_test).reshape(-1)

# Plotando gráfico dos valores de teste e dos valores previstos
plt.figure(figsize=(10,6))
plt.scatter(
    y_test,
    y_pred,
    color="blue",
    marker="o"
)
plt.plot(
    y_pred,
    y_pred,
    marker='o',
    color='green',
    markerfacecolor='red',
    markersize=7,
    linestyle='dashed'
)

plt.xlabel("Valores de teste")
plt.ylabel("Valores previstos")
plt.title("Valores previstos / teste")

plt.show()

# Visualizando a métrica R^2, quanto mais próximo de 1, melhor o ajuste do modelo aos dados
r2 = r2_score(y_test, y_pred)
print(f'R-squared (R²): {r2}')
```

# Conclusões sobre os modelos

Ao comparamos a regressão linear (R^2 aproximadamente 0.68) com a rede neural (R^2 aproximadamente 0.79), com ambas utilizando 100 épocas, podemos verificar que a rede neural conseguiu se adequar melhor aos dados por conta da quantidade de camadas e percéptrons que foram adicionados (normalizador e as camadas internas).  

Já a árvore de decisão, que obteve resultado de aproximadamente 0.70 com a métrica R^2 com profundidade 4, obteve overfit com profundidades maiores.

Logo, com esses algoritmos, entre eles, caso fossemos utilizar em produção, iríamos escolher a rede neural, pois ela pode se adequar melhor ao nosso caso. Além disso, ainda é possível tentar otimizar ainda mais, utilizando diferentes técnicas no pré-processamento ou testar mais parâmetros durante criação e treino dos modelos
