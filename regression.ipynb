# Dataset

O dataset escolhido foi o [House Sales in King County, USA](https://www.kaggle.com/datasets/harlfoxem/housesalesprediction/data). Ele possui preços de casas a venda entre os anos de 2014 e 2015 em King County nos estados unidos.

Iremos observar as colunas para entender o que cada uma significa, com a ajuda do fórum de discussões do Kaggle para as colunas menos intuitivas.

### Colunas

- **ID**: Identificador da casa
- **Date**: Data da venda
- **Price**: Preço da casa em dólares
- **Bedrooms**: Quantidade de quartos
- **Bathrooms**: Quantidade de banheiros (valores 0.5 significam lavabos)
- **Sqft_living**: Metros quadrados da casa
- **Sqft_lot**: Metros quadrados do lote
- **Floors**: Quantidade de andares da casa
- **Waterfront**: Diz se a casa está a beira-mar (0 e 1)
- **View**: Indica a qualidade da vista da propriedade, de 0 a 4.
- **Condition**: Condição atual da casa, de 1 a 5. Quanto maior, melhor conservada
- **Grade**:  Nota da casa após construção, de 1 a 13. Quanto maior, melhor
- **Sqft_above**: Metros quadrados considerando andares que estão acima do térreo
- **Sqft_basement**: Metros quadrados do porão
- **Yr_built**: Ano de construção
- **Yr_renovated**: Ano da última renovação
- **Zipcode**: CEP
- **Lat**: Latitude
- **Long**: Longitude
- **Sqft_living15**: Média de metros quadrados da casa dos 15 vizinhos mais próximos
- **Sqft_lot15**: Média de metros quadrados do lote dos 15 vizinhos mais próximos
